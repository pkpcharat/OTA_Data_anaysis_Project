```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


data = pd.read_excel('/content/FinalDE282-2.xlsx')

#ข้อมูลมีคอลัมน์ชื่อ 'Agoda', 'Traveloka', 'Trip'
agoda_prices = data['Agoda Price'].dropna()
traveloka_prices = data['Traveloka Price'].dropna()
booking_prices = data['booking Price'].dropna()


def plot_hist_with_normal(data, label, color, linestyle):
    mean = np.mean(data)
    std = np.std(data)
    plt.hist(data, bins=5, density=True, alpha=0.5, color=color, label=f'{label} Price')
    x = np.linspace(min(data), max(data), 100)
    plt.plot(x, norm.pdf(x, mean, std), linestyle, color=color, linewidth=2, label=f'{label} Normal Dist.')

# วาดกราฟ
plt.figure(figsize=(10, 6))
plot_hist_with_normal(agoda_prices, 'Agoda', 'blue', '--')
plot_hist_with_normal(traveloka_prices, 'Traveloka', 'orange', '-.')
plot_hist_with_normal(booking_prices, 'Booking', 'green', '-')

# การตั้งค่ากราฟ
plt.title('Histogram of OTAs Prices with Normal Distribution Curves')
plt.xlabel('Price')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
agoda_price_mean = data['Agoda Price'].mean()
agoda_price_std = data['Agoda Price'].std()
traveloka_price_mean = data['Traveloka Price'].mean()
traveloka_price_std = data['Traveloka Price'].std()
booking_price_mean = data['booking Price'].mean()
booking_price_std = data['booking Price'].std()

print(f"Agoda Price Mean: {agoda_price_mean}")
print(f"Agoda Price Standard Deviation: {agoda_price_std}")
print(f"Traveloka Price Mean: {traveloka_price_mean}")
print(f"Traveloka Price Standard Deviation: {traveloka_price_std}")
print(f"Booking Price Mean: {booking_price_mean}")
print(f"Booking Price Standard Deviation: {booking_price_std}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# List of prices
Agoda_prices = data['Agoda Price']

# Define bins starting from the minimum price
min_price = min(Agoda_prices)
max_price = max(Agoda_prices)
bins = np.linspace(min_price, max_price, 6)
# Plot the histogram
plt.figure(figsize=(10, 6))
hist, bin_edges, patches = plt.hist(Agoda_prices, bins=bins, color='royalblue', edgecolor='black', alpha=0.7)

# Add labels on the bars
for i in range(len(hist)):
    plt.text(bin_edges[i] + (bin_edges[1] - bin_edges[0]) / 2, hist[i] + 0.2,
             f'{hist[i]:.0f}', ha='center', color='white', fontsize=12, weight='bold')

# Customize x-axis tick labels to show bin ranges with decimals
bin_labels = [f'{bin_edges[i]:,.2f} - {bin_edges[i+1]:,.2f}' for i in range(len(bin_edges) - 1)]
plt.xticks(ticks=(bin_edges[:-1] + bin_edges[1:]) / 2, labels=bin_labels, rotation=45, fontsize=10)

# Customize the axes and titles
plt.title('Agoda Histogram', fontsize=16, weight='bold')
plt.xlabel('Price Range (THB)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from scipy.stats import shapiro

# Perform Shapiro-Wilk Test for Normality on Price Data

# Defining the columns(price columns)
price_columns = ['Agoda Price', 'Traveloka Price', 'booking Price']

# Iterate through each column and perform the test
print("Shapiro-Wilk Test Results:\n")
for column in price_columns:
   # Drop missing values for the column
   stat, p = shapiro(data[column].dropna())
   print(f"{column}:\n"
         f"  W-statistic = {stat:.3f}\n"
         f"  P-value = {p:.3f}")
   if p < 0.05:
       print(f"  -> Conclusion: Reject the null hypothesis. The data does not follow a normal distribution.\n")
   else:
       print(f"  -> Conclusion: Fail to reject the null hypothesis. The data appears to follow a normal distribution.\n")

```
